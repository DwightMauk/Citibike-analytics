```python
import pandas as pd
```

```python
df1 = pd.read_csv("JC-201807-citibike-tripdata.csv/JC-201807-citibike-tripdata.csv")
df1.head()
```

```python
df2 = pd.read_csv("JC-201808-citibike-tripdata.csv/JC-201808-citibike-tripdata.csv")
df2.head()
```

```python
df3 = pd.read_csv("JC-201809-citibike-tripdata.csv/JC-201809-citibike-tripdata.csv")
df3.head()
```

```python
citibike_2018Q3 = pd.concat([df1, df2, df3])
citibike_2018Q3
```

```python
citibike_2018Q3.drop(["starttime", "stoptime", "gender"], axis=1, inplace=True)
citibike_2018Q3
```

```python
citibike_2018Q3.to_csv("citibike_2018Q3.csv", index=False, encoding='utf8')
```
